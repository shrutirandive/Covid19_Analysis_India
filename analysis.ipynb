```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
covid_df=pd.read_csv('covid_19_india.csv')
covid_df
```

```python
covid_df.head()
```

```python
covid_df.isnull().sum()
```

```python
covid_df.describe()
```

```python
x=covid_df['Confirmed'].sum()
y=covid_df['Cured'].sum()
z=covid_df['Deaths'].sum()
active=x-y
print("total confirmed cases= ",x)
print("total cured cases= ",y)
print("total death cases= ",z)
print("total active cases= ",active)
```

```python
sns.barplot(x=['Confirmed', 'Cured', 'Deaths', 'active'], y=[x, y, z, active])
```

```python
#tail-->for new data i.e the last added 
covid_df1=covid_df.tail(10)
covid_df1
```

```python
covid_df1=covid_df1.sort_values(by="Confirmed", ascending=False).head(10)
covid_df1
```

```python
plt.figure(figsize=(16,5))
plt.bar(covid_df1['State/UnionTerritory'], height=covid_df1['Confirmed'], color='blue')
plt.xlabel('State/union terroritory')
plt.ylabel('confirmed cases')
plt.title('Confirmed cases',size=20)
sns.set()      #displays grid on graph i.e horizontal vertical lines, also refer https://www.mygreatlearning.com/blog/seaborn-tutorial/
plt.show()
```

```python
covid_df1['Cure-Ratio']=covid_df1['Cured']/covid_df1['Confirmed']
```

```python
plt.figure(figsize=(20, 5))
plt.title('Cure Ratio of Top 10 Contaminated States',size=20)
sns.pointplot(data=covid_df1, x='State/UnionTerritory', y='Cure-Ratio', color="red")
sns.set()
plt.show()
```

```python
covid_df1['Fatality-ratio']=covid_df1['Deaths']/covid_df1['Confirmed']
```

```python
plt.figure(figsize=(20,5))
plt.title('Fatality-Ratio of Top 10 Contaminated States',size=20)
sns.pointplot(data=covid_df1,x='State/UnionTerritory',y='Fatality-ratio',color='green')
sns.set_style("whitegrid")
plt.show()
```
