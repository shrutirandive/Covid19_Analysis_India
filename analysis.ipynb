```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid_df=pd.read_csv('covid_19_india.csv')
covid_df
```

```python
covid_df.head()
```

```python
covid_df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
plt.plot(covid_df['Confirmed'])
plt.title('confirmed cases')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(covid_df['Deaths'])
plt.title('death cases')
plt.show()
```

```python
covid_df['State/UnionTerritory'].unique()
```

```python
covid_df.describe()
```

```python
x=covid_df['Confirmed'].sum()
y=covid_df['Cured'].sum()
z=covid_df['Deaths'].sum()
active=x-y
print("total confirmed cases= ",x)
print("total cured cases= ",y)
print("total death cases= ",z)
print("total active cases= ",active)
```

```python
sns.barplot(x=['Confirmed', 'Cured', 'Deaths', 'active'], y=[x, y, z, active])
```

```python
covid_df1=covid_df.tail(35)
covid_df1
```

```python
covid_df2=covid_df1.sort_values(by="Confirmed", ascending=False).head(10)
covid_df2
```

```python
plt.figure(figsize=(16,5))
plt.bar(covid_df2['State/UnionTerritory'], height=covid_df2['Confirmed'], color='blue')
plt.xlabel('State/union terroritory')
plt.ylabel('confirmed cases')
```

```python
covid_df2['Cure-Ratio']=covid_df2['Cured']/covid_df['Confirmed']
```

```python
plt.figure(figsize=(20, 5))
plt.title('cure ratio')
sns.pointplot(data=covid_df2, x='State/UnionTerritory', y='Cure-Ratio', color="red")
```

```python
covid_df2['Fatality-ratio']=covid_df2['Deaths']/covid_df2['Confirmed']
```

```python
plt.figure(figsize=(20,10))
plt.title('Fatality-Ratio of Top 10 Contaminated States',size=15)
sns.pointplot(data=covid_df2,x='State/UnionTerritory',y='Fatality-ratio',color='green')
plt.show()
```

```python
# for Maharashtra
maha_df=covid_df[covid_df['State/UnionTerritory']=='Maharashtra']
maha_df
```

```python
#data filtering
#as we want a time series so that we have decided any date and filter that date data only

date_df=maha_df[maha_df['Date'].isin(['2020-03-11','2020-04-11','2020-05-11', '2020-06-11', '2020-07-11', '2020-08-11',
                                      '2020-09-11','2020-10-11','2020-11-11', '2020-12-11','2021-01-11','2021-02-11',
                                      '2021-03-11', '2021-04-11'])]
```

```python
date_df
```

```python
plt.figure(figsize=(20,7))
sns.pointplot(data=date_df,x='Date',y='Confirmed',color='Black')
sns.pointplot(data=date_df,x='Date',y='Cured',color='yellow')
sns.pointplot(data=date_df,x='Date',y='Deaths',color='Red')
plt.show()
```

```python
#plot for all states
date_df2=covid_df[covid_df['Date'].isin(['2020-03-11','2020-04-11','2020-05-11', '2020-06-11', '2020-07-11', '2020-08-11',
                                      '2020-09-11','2020-10-11','2020-11-11', '2020-12-11','2021-01-11','2021-02-11',
                                      '2021-03-11', '2021-04-11'])]
date_df2
```

```python
plt.figure(figsize=(16,10))
sns.pointplot(data=date_df2,x='Date',y='Confirmed',hue='State/UnionTerritory')
plt.legend(bbox_to_anchor=(1,1))
plt.title("CONFIRM CASES")
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.pointplot(data=date_df2,x='Date',y='Deaths',hue='State/UnionTerritory')
plt.legend(bbox_to_anchor=(1,1))
plt.title("DEATH CASES")
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.pointplot(data=date_df2,x='Date',y='Cured',hue='State/UnionTerritory')
plt.legend(bbox_to_anchor=(1,1))
plt.title("CURED CASES")
plt.show()
```
